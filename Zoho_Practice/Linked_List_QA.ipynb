## 1. pattern: Two-Pointer (Fast and Slow)
This pattern is ideal for detecting cycles or finding nodes relative to the end of the list.

### Linked list cycle

```python
from Single_Linked_List import LinkedList, Node

def checkCycle(head: Node) -> bool:
    slow = head
    fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False


values = [1,2,3,4,5]
ll = LinkedList()
for val in values:
    ll.insertAtEnd(val)
ll.traverse()
print(checkCycle(ll.head))   
ll.cycleList()
print(checkCycle(ll.head))
# ll.traverse()
```

## 2. Pattern: In-place Reversal
Essential for problems involving rearranging nodes without allocating new memory.

```python
from Single_Linked_List import LinkedList, Node

def reverseList(head: Node) -> Node:
    curdNode = head
    prevNode = None
    while curdNode:
        temp = curdNode.next
        curdNode.next = prevNode
        prevNode = curdNode
        curdNode = temp
        
    return prevNode

arr = [1,2,3,4,5,6,7]
ll = LinkedList()
for a in arr:
    ll.insertAtEnd(a)
node = reverseList(ll.head)
while node:
    print(node.data)
    node = node.next
```

## 3.Pattern: Using dummy nodes
The "Dummy Node" (or Sentinel Node) is one of the most powerful design patterns in linked list manipulation. Its primary purpose is to simplify code by providing a stable reference point that remains fixed throughout the operation, effectively turning "head-specific" edge cases into standard iterative logic.

### Merging Two Sorted Linked Lists


```python
from Single_Linked_List import LinkedList, Node
def mergeTwoSortedList(l1: Node, l2: Node) -> Node:
    dummy = Node(0)
    tail = dummy
    while l1 and l2:
        if l1.data < l2.data:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next
        
    tail.next = l1 or l2
    return dummy.next

ll1 = LinkedList()
ll2 = LinkedList()
for val in [1,3,5]:
    ll1.insertAtEnd(val)
for val in [2,4,6]:
    ll2.insertAtEnd(val)
node = mergeTwoSortedList(ll1.head, ll2.head)
while node:
    print(node.data)
    node = node.next
    
```

### Partition List
Given the head of a linked list and a value x, partition it such that all nodes less than x come before nodes greater than or equal to x.

```python
from Single_Linked_List import LinkedList, Node

def partition(head: Node, x: int) -> Node:
    dummy = Node(0)
    dummy.next = head
    prev = dummy
    left = head
    right = head
    while right:
        if right.data < x and right != left:
            prev.next = right.next
            right.next = left.next
            left.next = right
            right = prev.next
        else:
            right = right.next
    return dummy.next
            
ll = LinkedList()
for val in [1,4,3,2,5,2]:
    ll.insertAtEnd(val)
    
node = partition(ll.head, 3)
while node:
    print(node.data)
    node = node.next


            
            
            
            
```
