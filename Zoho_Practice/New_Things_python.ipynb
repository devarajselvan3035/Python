The best and most Pythonic way to find common key-value pairs between two dictionaries is to use the & (intersection) operator on their .items() views, as documented 

```python
dict_a = {'x': 1, 'y': 2, 'z': 3}
dict_b = {'w': 0, 'x': 1, 'y': 9}
dict_c = {'x': 1}

# 1. Find the intersection (returns a set of tuples)
common_set = dict_a.items() & dict_b.items() & dict_c.items()

# 2. Convert back to a dictionary
common_dict = dict(common_set)

print(common_dict)  # Output: {'x': 1}
```

```python
dict_a = {'x': [1], 'y': [2]}
dict_b = {'x': [1], 'y': [5]}

# Safe for unhashable values like lists
common_dict = {k: v for k, v in dict_a.items() if k in dict_b and dict_b[k] == v}

print(common_dict)  # Output: {'x': [1]}
```

backtracking

```python
def subsets(nums):
    result = []
    def backtracking(index, path):
        if index == len(nums):
            result.append(path.copy())
            return
       
        path.append(nums[index]) 
        backtracking(index+1, path)
        path.pop()
        
        backtracking(index+1, path)
    backtracking(0, [])
    return result
    
nums = [1,2,3,4]
print(subsets(nums))
```

combinations

```python
def combination(nums, k):
    result = []
    def backtracking(index, arr):
        if len(arr) == k:
            result.append(arr.copy())
            
        for i in range(index, len(nums)):
            arr.append(nums[i])
            backtracking(i+1, arr)
            arr.pop()
            
    backtracking(0, [])
    return result

nums, k = [1,2,3,4], 2
print(combination(nums, k))
```

permutation

```python
def permutation(nums):
    result = []
    def backtracking(idx, arr):
        print(idx)
        if idx == len(nums):
            if len(arr) == len(nums):
                result.append(arr.copy())
            return
        
        for i in range(idx, len(nums)):
            arr.append(nums[i])
            backtracking(idx+1, arr)
            arr.pop()
        
            backtracking(idx+1, arr)
    
    for i in range(len(nums)):
        backtracking(i, [])
    return result
    
nums = [1,2,3]
print(permutation(nums))
```
