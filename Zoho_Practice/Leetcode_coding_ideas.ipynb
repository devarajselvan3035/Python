350. Intersection of two arrays

```python
def intersection(arr1: list[int], arr2: list[int]) -> list[int]:
    arr1.sort()
    arr2.sort()
    l, r = 0, 0
    res = []
    idx = 0
    while l < len(arr1) and r < len(arr2):
        if arr1[l] == arr2[r]:
            res.append(arr1[l])
            l += 1
            r += 1
        elif arr1[l] < arr2[r]:
            l += 1
        elif arr2[r] < arr1[l]:
            r += 1
    return res


num1, num2 = [1, 2, 3, 1], [1, 3, 2, 1]
print(intersection(num1, num2))
```

```python
def thirdMaxNum(arr: list[int]) -> int:
    arr = set(arr)
    first, second, third = float('-inf'), float('-inf'), float('-inf')
    for val in arr:
        first, second, third = second, third, max(third, val)
    return first if first != float("-inf") else max(arr)
    
arr = [2,2,4,1]
arr = [2,2,2,1,1,1,3]
print(thirdMaxNum(arr))
```

496. Next Greater Element I

```python
def nextGreatElement(nums1: list[int], nums2: list[int]) -> list[int]:
    nge = ngeGenerate(nums2)
    ans = []
    for num in nums1:
        idx = nums2.index(num)
        ans.append(nge[idx])
    return ans
        
    
def ngeGenerate(arr: list[int]) -> list[int]:
    ans = []
    stack = []
    for val in arr[::-1]:
        while stack and val >= stack[-1]:
            stack.pop()
        if not stack:
            ans = [-1] + ans
        else:
            ans = [stack[-1]] + ans
        stack.append(val)
    return ans

nums1, nums2 = [4,1,2], [1,3,4,2]
print(nextGreatElement(nums1, nums2))
```

```python
def arrayPartition(nums: list[int]) -> int:
    nums.sort(reverse=True)
    ans = 0
    for i in range(0, len(nums), 2):
        ans += min(nums[i], nums[i+1])
    return ans
        
nums = [6,2,6,5,1,2]
# arrayPartition(nums)
nums.sort(reverse=True)
nums[::2]
    
```

599. Minimum Index Sum of Two Lists

```python
def minIndex(list1: list[str], list2: list[str]) -> str:
    ans = [float('inf'), []]
    for idx, val in enumerate(list1):
        if val in list2:
            idy = list2.index(val)
            if ans[0] == abs(idx+idy):
                ans[-1].append(val)
            ans = min(ans, [abs(idx+idy), ans[-1] + [val]])
    return ans
            
    

list1 = ["Shogun","Tapioca Express","Burger King","KFC"]
list2 = ["Piatti","The Grill at Torrey Pines","Hungry Hunter Steakhouse","Shogun"]
# list1 = ["happy","sad","good"]
# list2 = ["sad","happy","good"]

print(minIndex(list1, list2))
```

1122. Relative Sort Array

```python
def relativeSort(arr1: list[int], arr2: list[int]) -> list[int]:
    arr_dict = {v:i for i, v in enumerate(arr2)}
    
    # def sortKey(val: int, d: dict):
    #     if val in d:
    #         idx = d[val]
    #         return idx
    #     else:
    #         return len(d) + 1
    
    have = [v for v in arr1 if v in arr2] 
    not_have = [v for v in arr1 if v not in arr2] 
    
    have.sort(key=lambda x: arr_dict[x])
    return have + not_have

arr1, arr2 = [2,3,1,3,2,4,6,7,9,2,19], [2,1,4,3,9,6]
arr1, arr2 = [28,6,22,8,44,17], [22,28,8,6]
print(relativeSort(arr1, arr2))
```

1160. Find Words That Can Be Formed By Characters

```python
def wordCount(words: list[str], chars: str) -> int:
    chars = set(chars)
    res = 0
    for word in words:
        if set(word) < chars:
            res += len(word)
    return res 
    
words = ["cat","bt","hat","tree"]
chars = "atach"
words = ["hello","world","leetcode"]
chars = "welldonehoneyr"
print(wordCount(words, chars))
```

1346. Check If N and Its Double Exist

```python
def checkIfExist(arr: list[int]) -> bool:
    res_dict = {}
    for v in arr:
        if v in res_dict and (res_dict[v] == v/2 or res_dict[v] == v*2):
            return True
        res_dict[v/2] = v
        res_dict[v*2] = v
    return False

arr = [10,2,5,3]
# arr = [3,1,7,11]
print(checkIfExist(arr))
```

```python
def fillCups(amount: List[int]) -> int:
    amount.sort(reverse=True)
    if amount[0] >= amount[1] + amount[2]:
        return amount[0]
    return sum(amount) // 2 + sum(amount) % 2
arr = [5,4,4]
print(fillCups(arr))
```

1475. Final Prices with a Special Discount in a Shop


```python
def discountPrice(arr: list[int]) -> list[int]:
    stack = []
    ans = []
    for val in arr[::-1]:
        while stack and val < stack[-1]:
            stack.pop()
        if not stack:
            ans = [val] + ans
        else:
            ans = [val - stack[-1]] + ans
        stack.append(val)
    return ans
    
prices = [8,4,6,2,3]
print(discountPrice(prices))
```

402. Remove K Digits


```python
def removeKDigits(num: str, k: int) -> int:
    list_digits = list(num)
    l, r = 0, 1
    while k > 0 and r < len(list_digits):
        if list_digits[r] > list_digits[l]:
            list_digits[r], list_digits[l] = list_digits[l], list_digits[r]
            l += 1
        elif list_digits[r] == '0':
            l += 2
        r += 1
        k -= 1
    ans  = list_digits[l:] if r != len(list_digits) and k <= 0 else list_digits[r:]
    return "".join(ans) if ans else '0'

num, k = "1432219", 3
num, k = "10200", 1
num, k = "10", 2
print(removeKDigits(num, k))
```

581. Shortest Unsorted Continuous Subarray (***)(SORT)


```python
def shortestUnsorted(arr: list[int]) -> int:
    left, right = 0, 0
    sorted_arr = sorted(arr)
    
    for idx in range(len(arr)):
        if arr[idx] != sorted_arr[idx]:
            if left == 0:
                left = idx
            right = idx+1
            
    return right - left
            
nums = [2,6,4,8,10,9,15]
print(shortestUnsorted(nums))
```

71. Simplify Path

```python
def simplifyPath(path: str) -> str:
    stack = []
    split_path = [p for p in path.split('/') if p]
    for path in split_path:
        if path == '..':
            stack.pop()
        elif path == '.':
            continue
        else: 
            stack.append(path)
    return '/' if not stack else "/" + "/".join(stack)
        
    
    
path = "/home/"
# path = "/home/user/Documents/../Pictures"
# path = "/.../a/../b/c/../d/./"
path = "/../"
print(simplifyPath(path))
```
