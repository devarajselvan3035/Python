## String Manipulation & Custom parsing

**Beginner**: Given a sentence, remove all words that are palindromes and return the modified stirng (Examples: "He did a good deed" -> "He good")

````python
def checkPalindrome(s: str) -> bool:
    s = s.lower()
    l, r = 0, len(s)-1
    while l <= r:
        if s[l] != s[r]:
            return False
        l += 1
        r -= 1
    return True

def removePalindrome(s: str) -> str:
    word_list = s.split(" ")
    res = []
    for word in word_list:
        if not checkPalindrome(word):
            res.append(word)
    return " ".join(res)

print(removePalindrome("He did a good deed"))
        
````

**Intermediate**: Govem a compressed input string like `a1b10c2` expand it to output `abbbbbbbbbbcc`. The repetition factor can be up to 2 digits.

````python
def stringExpansion(string: str) -> str:
    string = string + 'a'
    s: str = ''
    i: str = '0' 
    res: str = ""
    for c in string:
        if c.isalpha():
            if i != '':
                res += s * int(i)
                s = ''
            s += c
            i = ''
        else:
            i += c
    return res
print(stringExpansion("a1b10c2"))
````

**Intermediate**: Expands a string with encoded repeat counts inside brackets. \
    Example: '3[a]2[bc]' -> 'aaabcbc' \
             '2[a3[b]]'  -> 'abbbabbb' 

````python
def stringExpansion2(string: str) -> str:
    stack = []
    curd_num = ''
    curd_str = ''
    for chr in string:
        if chr.isdigit():
            curd_num += chr
        elif chr == '[':
            stack.append((curd_str, curd_num))
            curd_num = ''
            curd_str = ''
        elif chr == ']':
            prev_str, num = stack.pop()
            curd_str = prev_str + (curd_str * int(num))
        else:
            curd_str += chr
    return curd_str
print(stringExpansion2("2[a3[b]]"))
````

You are not allowed to use high-level library helper functions like `split()`, `join()`, or slicing steps that reverse the string implicitly. The parsing must proceed word-by-word, and the accumulation must be managed on the stack frame of a recursive call.

- **Input Sentence**: `"I love india"` 
- **Expected Output**: `"india love I"` 

````python
def wordReversal(s: str) -> str:
    s = s + " "
    word = ''
    word_list = []
    for chr in s:
        if chr == " ":
            word_list.append(word)
            word = ''
        else:
            word += chr
    return word_list[::-1]

s = "I Love India"
print(wordReversal(s))
        
````

**Beginner**: Seperate 0s and 1s in an array in a single traversal pass using 0(1) extra space.

Input: `[1,0,1,0,1,0]`
Output: `[0,0,0,1,1,1]`

````python
def seperate(arr: list[int]) -> list[int]:
    l = 0
    for r in range(len(arr)):
        if arr[r] == 0:
            arr[r], arr[l] = arr[l], arr[r]
            l += 1
    return arr

print(seperate([1,0,1,0,1,0]))
            
````

```
Input: arr = [3, 1, 4, 6, 2, 5]
Output: [6, 1, 5, 2, 4, 3]
```

````python
def alterMinMaxSort(arr: list[int]) -> list[int]:
    for l in range(len(arr)):
        min_value = [float('inf'), 0]
        max_value = [float('-inf'), 0]
        for r in range(l, len(arr)):
            if l % 2 == 0:
                if arr[r] > max_value[0]:
                    max_value[0] = arr[r]
                    max_value[1] = r
                if r == len(arr)-1:
                    arr[l], arr[max_value[1]] = arr[max_value[1]], arr[l]
            else:
                if arr[r] < min_value[0]:
                    min_value[0] = arr[r]
                    min_value[1] = r
                if r == len(arr)-1:
                    arr[l], arr[min_value[1]] = arr[min_value[1]], arr[l]
    return arr

arr = [3,1,4,6,2,5]
print(alterMinMaxSort(arr))
````

Given an array of positive integers, sort the array in descending order based on the count of distinct factors of each element. If two elements have the same number of factors, preserve their relative position from the original input array to ensure a stable sort.
```
Input: arr = [5, 12, 6]
Output: [12, 6, 5]

Explanation: 
Let's calculate the distinct factors for each number:12: 
Factors are 1, 2, 3, 4, 6, 12 $\rightarrow$ 6 
factors6: Factors are 1, 2, 3, 6 $\rightarrow$ 4 
factors5: Factors are 1, 5 $\rightarrow$ 2 factorsSorting by factor count descending gives us 12 (6 factors), then 6 (4 factors), then 5 (2 factors).
```

````python
def factorCount(val: int) -> int:
    if val == 1:
        return 1
    count = 2
    for i in range(2, val // 2+1):
        if val % i == 0:
            count += 1
    return count

def factorSort(arr: list[int]) -> list[int]:
    arr.sort(key=factorCount, reverse=True)
    return arr

arr = [5, 12, 6]
arr = [1,2,3,4,5,6,7,8]
print(factorSort(arr))
````

**High**: Given an array of integers, sort the elements at odd positions (using 1=based indexing) in descending order, and sort the elements at even positions in ascending order.

Input: [3, 1, 2, 4, 5, 8]
Output: [5, 1, 3, 4, 2, 8]

````python
def positionSort(arr: list[int]) -> list[int]:
    odd_position = []
    even_position = []
    for i, v in enumerate(arr):
        if i%2 == 0:
            even_position.append(v)
        else:
            odd_position.append(v)
            
    odd_position.sort()
    even_position.sort(reverse=True)
   
    odd_idx, even_idx = 0, 0 
    for i in range(len(arr)):
        if i%2 != 0:
            arr[i] = odd_position[odd_idx]
            odd_idx += 1
        else:
            arr[i] = even_position[even_idx]
            even_idx += 1
    return arr
arr = [3,1,2,4,5,8]
print(positionSort(arr))
````

**Begineeer**: Write a function to check if a provided number is an armstrong number (equal to the sum of its own digits each raised to the power of the total number of digits).

````python
def armstrong(n: int) -> bool:
    original = n
    res = 0
    length = len(str(n))
    while n > 0:
        rem = n % 10
        res += rem ** length
        n //= 10
    return True if original == res else False
print(armstrong(123))
````

````python
def isLeapYear(year: int) -> bool:
    if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
        return True
    return False

def totalDays(date1: list[int], date2: list[int]) -> int:
    day1, month1, year1 = date1
    day2, month2, year2 = date2
    month1, month2 = month1-1, month2-1
    total_sum = 0
    months_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30 , 31]
    for year in range(year1, year2+1):
        if isLeapYear(year):
            months_days[1] = 29
        total_sum += sum(months_days)
 
    if isLeapYear(year1):
        months_days[1] == 29
    for month in months_days[:month1]:
        total_sum -= month
    total_sum -= day1
    
    if isLeapYear(year2):
        months_days[1] == 29
    for month in months_days[month2+1:]:
        total_sum -= month
    total_sum -= months_days[month2] - day2
    
    return total_sum

date1 = [28, 2, 2000]
date2 = [1, 3, 2000]
print(totalDays(date1, date2))
        
    
            
````

**Intermediate**: Given a sequence of digits (e.g., "121"), count the number of possible alphabetical decodings if 1 represents 'A', 2 represents 'B' up to 26 for 'z'

````python
def num_decodings(s: str) -> int:
    """
    Calculates the total number of ways to decode a digit string 
    into letters ('1' = 'A', ..., '26' = 'Z').
    """
    # An empty string or a string starting with '0' cannot be decoded
    if not s or s[0] == '0':
        return 0

    n = len(s)
    
    # dp[i] will store the number of ways to decode a prefix of length i
    dp = [0] * (n + 1)
    print(dp) 
    # Base Cases:
    # An empty prefix has 1 valid configuration path
    dp[0] = 1
    # A single-digit prefix (that isn't '0') has 1 valid path
    dp[1] = 1

    # Build the solution for longer prefixes
    for i in range(2, n + 1):
        # 1. Look at the single digit at index i-1
        single_digit = int(s[i-1:i])
        if single_digit != 0:
            dp[i] += dp[i-1]
        print(f"dp : {dp}")

        # 2. Look at the two-digit combination ending at index i-1
        two_digits = int(s[i-2:i])
        if 10 <= two_digits <= 26:
            dp[i] += dp[i-2]

    return dp[n]

# --- Example Usage ---
print(num_decodings("121"))   # Output: 3  (ABA, LA, AU)
print(num_decodings("12"))    # Output: 2  (AB, L)
print(num_decodings("226"))   # Output: 3  (BBF, BZ, VF)
print(num_decodings("06"))    # Output: 0  (Invalid sequence, starts with 0)
print(num_decodings("1111"))    # Output: 1  (J)
````

**Intermediate**: Implement an expression validator that ensures every parenthesis type `(` has a matching closing `)` and that brackets are nested correctly.
```
Input: s = "()[]{}"
Output: True

Input: s = "{[()]}"
Output: True

Input: s = "([)]"
Output: False
```

````python
def parenthesisValidation(parathesis: str) -> bool:
    paraDict = {')':'(', ']':'[', '}':'{'}
    stack = []
    for chr in parathesis:
        if chr in "([{":
            stack.append(chr)
        else:
            if paraDict[chr] == stack[-1]:
                stack.pop()
    return stack

# s = "(){}[]"
s = "({[]})"
print(parenthesisValidation(s))
            
````
