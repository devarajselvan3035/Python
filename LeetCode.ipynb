## Roman to Integer
Roman numerals are represented by seven different symbols:I, V, X, L, C, D and M.\
**Examples**\
Inputs: "MCMXCIV"\
Outputs: 1994\
Explanation: M = 1000, CM = 900, XC = 90 and IV = 4.

```python
roman_dict = {'I':1, 'V':5, 'X':10, 'L':50, 'C':100, 'D':500, 'M':1000}
Inputs = "MCMXCI"
def find_roman(roman):
    ans = 0
    for i in range(len(roman)-1):
        if roman_dict[roman[i]] < roman_dict[roman[i+1]]:
           ans -= roman_dict[roman[i]] 
        else:
            ans += roman_dict[roman[i]]
    return ans + roman_dict[roman[-1]]
print(find_roman(Inputs))
```

## 26. Remove Duplicates from sorted array

Input: nums = [1,1,2]\
Outputs: 2, nums = [1,2,_]

Inputs: nums = [0,0,1,1,1,2,2,3,3,4]\
Outputs: 5, nums = [0,1,2,3,4,_,_,_,_,_]


```python
nums = [0,0,1,1,1,2,2,3,3,4]

def removeDuplicates(nums:list) -> int:
    l, r = 0, 0
    while r < len(nums):
        if nums[l] != nums[r]:
            nums[l+1], nums[r] = nums[r], nums[l+1]
            l += 1

        r += 1
    print(l)
    return nums
print(removeDuplicates(nums))
```

## 27. Remove Element
Inputs: nums = [3,2,2,3], val = 3\
Outputs: 2, nums = [2,2,_,_]

Inputs: nums = [0,1,2,2,3,0,4,2], val=2\
outputs: 5, nums = [0,1,4,0,3]

```python
nums = [0,1,2,2,3,0,4,2]

def removeElement(nums:list, val:int) -> int:
    l, r = 0, len(nums)-1
    while l < r:
        if nums[r] == val:
            r -= 1
        elif nums[l] == val:
            nums[l], nums[r] = nums[r], nums[l]
        else:
            l += 1
        print(nums, nums[l], nums[r])
    print(l)
    return nums

print(removeElement(nums, 2))
```
