## Reverse String using Recursion
ip = hello world\
op = dlrow olleh

```python
def reverse(string: str) -> str:
    if len(string) == 1:
        return string[0]
    return reverse(string[1:]) + string[0]


# print(reverse("hello world"))
```

## check Palindrome using Recursion
Check given ip and reverced ip both are same

ip = kayak\
op = True

```python
def Palindrome(string: str) -> bool:
    if len(string) == 1 or len(string) == 0:
        return True
    if string[0] == string[-1]:
        return Palindrome(string[1:-1])
    return False


# print(Palindrome("raccar"))
```

## Decimal to Binary
ip = 233\
op = 11101001

```python
def decimalTobinary(dec: int) -> str:
    if dec <= 1:
        return str(dec)
    return decimalTobinary(dec // 2) + str(dec % 2)


print(decimalTobinary(233))
```

## Sum of Natural Numbers

ip = 5\
op = (1+2+3+4+5) =15

```python
def sumOfNaturalnum(num:int) -> int:
    if num == 1:
        return 1
    return num + sumOfNaturalnum(num-1)
print(sumOfNaturalnum(5))
```

## Binary Search

It returns bool values. If `x` value inside the `arr` return `True`, else return `False`


```python
def binarySearch(arr:list, x:int) -> bool:
    if len(arr) == 0:
        return False
    if len(arr) == 1:
        return True if arr[0] == x else False
    print(arr)
    mid = len(arr) // 2
    print(mid)
    if x == arr[mid]:
        return True
    elif x < arr[mid]:
        return binarySearch(arr[:mid], x)
    else:
        return binarySearch(arr[mid+1:], x)
    return False
print(binarySearch([1,2,3,4,5,6], 10))
```

## Binary Search
It returns position of the given `x` value.

```python
def binarySearchPos(arr:list, left:int, right:int, x:int):
    if left > right:
        return -1
    mid = (left+right) // 2
    if arr[mid] == x:
        return mid
    elif x < arr[mid]:
        binarySearchPos(arr, left, mid-1, x)
    else:
        binarySearchPos(arr, mid+1, right, x)

arr = [1,2,3,4,5]
print(binarySearchPos(arr, 0, len(arr)-1, 3))
        
```

## Fibonacci
ip = 5 \
op = 0+1+1+2+3+5 = 12

```python
def fibonacci(num:int) -> int:
    if num == 0 or num == 1:
        return num
    return fibonacci(num-1) + fibonacci(num-2)
print(fibonacci(6))
```
