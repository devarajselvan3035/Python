## Pair Sum - Unsorted
Given an array of integers, return the indexes of any two numbers that add up to a target. The order of the indexes in the result doesn't matter. If no pair is found, return an empty array.

**Example**\
Inputs: nums = [-1, 3, 4, 2]\
Output: [0, 2]\
> The same index cannot be user twice in the result

```python
from typing import List

nums = [-1, 3, 4, 2]

def pair_sum(arr:List[int], target:int):
    hash_map = {}
    for i, n in enumerate(arr):
        if target-n in hash_map:
            return [i, hash_map[target-n]]
        else:
            hash_map[n] = i
    return []

print(pair_sum(nums, 3))
        
    
```

## Verify Sudoku Board
Given a partially completed 9x9 sudoku board, determine if the current state of the board adheres to the rules of the game
- Each row and column must contain unique numbers between 1 to 0, or be empty (represented as 0).
- Each of the nine 3x3 subgrids that compose the grid must contain unique numbers between 1 and 9 or be empty.

Output: False
> Assume each integer on the board falls in the range of [0, 9]

```python
board = [["8","3",".",".","7",".",".",".","."]
,["6",".",".","1","9","5",".",".","."]
,[".","9","8",".",".",".",".","6","."]
,["8",".",".",".","6",".",".",".","3"]
,["4",".",".","8",".","3",".",".","1"]
,["7",".",".",".","2",".",".",".","6"]
,[".","6",".",".",".",".","2","8","."]
,[".",".",".","4","1","9",".",".","5"]
,[".",".",".",".","8",".",".","7","9"]]

def validate_sudoku(arr):
    hash_column = [set() for _ in range(9)]
    hash_row = [set() for _ in range(9)]
    hash_grid = [[set() for _ in range(34)] for _ in range(3)]
    
    for c in range(9):
        for r in range(9):
            if arr[c][r] in hash_column[c]:
                return False
            elif arr[c][r] in hash_row[r]:
                return False
            elif arr[c][r] in hash_grid[r//3][c//3]:
                return False
            elif arr[c][r] != '.':
                hash_column[c].add(arr[c][r])
                hash_row[r].add(arr[c][r])
                hash_grid[r//3][c//3].add(arr[c][r])
    return True

print(validate_sudoku(board))
```

## Zero Striping
For each zero in an mxn matrix, set its entire row and column to zero in place

Input: matrix = [[0,1,2,0],[3,4,5,2],[1,3,1,5]]\
Output: [[0,0,0,0],[0,4,5,0],[0,3,1,0]]

```python
matrix = [[0,1,2,0],[3,4,5,2],[1,3,1,5]]

row = set()
col = set()

for c in range(len(matrix)):
    for r in range(len(matrix[0])):
        if matrix[c][r] == 0:
            row.add(r)
            col.add(c)
            
for c in range(len(matrix)):
    for r in range(len(matrix[0])):
        if r in row:
            matrix[c][r] = 0
        elif c in col:
            matrix[c][r] = 0
            
print(matrix)
```
